```python
import requests
import IPython
import ffmpeg
import re
```

```python
url = "http://localhost:6001/tts"
```

```python
def text2speech(text: str):
    data = {"text": text}
    resp = requests.post(url, json=data, headers = {"Content-Type": "application/json"})
    return resp.content

def split_sentences(paragraph):
    sentences = []
    sentence = ""
    for char in paragraph:
        sentence += char
        if char in [".", "!", "?"]:
            sentences.append(sentence.strip())
            sentence = ""
    if sentence:
        sentences.append(sentence.strip())
    return sentences

def concat(files: list, output: str):
    input_streams = []
    for file in files:
        input_streams.append(ffmpeg.input(file))
    output = ffmpeg.concat(*input_streams, v=0, a=1).output(output)
    ffmpeg.run(output)
```
